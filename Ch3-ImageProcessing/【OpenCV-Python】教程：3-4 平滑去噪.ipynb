```python
import cv2
import numpy as np


def addPepperSalt(src: cv2.Mat, snr: float=0.99)  -> cv2.Mat : 
    """
    addPepperSalt 给彩色图像添加黑白椒盐噪声

    _extended_summary_

    Args:
        src (cv2.Mat): 输入图像
        snr (float): 信噪比
    """
    if src is None:
        return None
    
    dst = src.copy()
    
    h, w, _ = src.shape
    imgsz = h * w 
    pepperSaltNum = int(imgsz * (1.0 - snr))
    
    for i in range(pepperSaltNum):
        randx = np.random.randint(1, w-1)
        randy = np.random.randint(1, h-1)
        
        if np.random.random() <= 0.5:
            dst[randy, randx] = 0
        else:
            dst[randy, randx] = 255            
    return dst


img = cv2.imread('assets/opencv-logo-white.png')
cv2.imshow("src", img)

pepperSaltIm = addPepperSalt(img, 0.95)
median = cv2.medianBlur(pepperSaltIm, 5)
cv2.imshow("peppersalt", pepperSaltIm)
cv2.imshow("median", median)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import numpy as np
import cv2

def addPepperSalt(src: cv2.Mat, snr: float = 0.99) -> cv2.Mat:
    """
    addPepperSalt 给彩色图像添加黑白椒盐噪声

    _extended_summary_

    Args:
        src (cv2.Mat): 输入图像
        snr (float): 信噪比
    """
    if src is None:
        return None

    dst = src.copy()

    h, w, _ = src.shape
    imgsz = h * w
    pepperSaltNum = int(imgsz * (1.0 - snr))

    for i in range(pepperSaltNum):
        randx = np.random.randint(1, w-1)
        randy = np.random.randint(1, h-1)

        if np.random.random() <= 0.5:
            dst[randy, randx] = 0
        else:
            dst[randy, randx] = 255
    return dst


img = cv2.imread('assets/bilateral.jpeg')
# img = cv2.imread('assets/sheets.png')
cv2.imshow("src", img)

pepperSaltIm = addPepperSalt(img, 0.99)
cv2.imshow("peppersalt", pepperSaltIm)

median = cv2.medianBlur(pepperSaltIm, 5)
blur = cv2.bilateralFilter(pepperSaltIm, 15, 180, 180)

cv2.imshow("median", median)
cv2.imshow("bilateralFilter", blur)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
